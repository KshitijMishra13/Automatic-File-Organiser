```python
import os, shutil, time
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler
```

```python
source = r"ENTER SOURCE FOLDER PATH HERE"
```

```python
file_types = {
    ".pdf": "PDF Files",
    ".xlsx": "Excel Files",
    ".zip": "ZIP Files",
    ".jpg": "Images",
    ".png": "Images",
    ".txt": "Text Files",
    ".mkv": "Videos",
    ".mp4": "Videos",
    ".avi": "Videos",
    ".mov": "Videos",
    ".docx": "Word Files",
    ".doc": "Word Files",
    ".pptx": "PowerPoint Files",
    ".ppt": "PowerPoint Files",
    ".csv": "CSV Files",
    ".json": "JSON Files"
}
```

```python
default_folder = "Other Files"
```

```python
# Create folders if they don't exist
def create_folders(folder_path):
    if not os.path.exists(folder_path):
        os.mkdir(folder_path)
```

```python
def move_file(file_path):
    if os.path.isdir(file_path):
        return
    file_name = os.path.basename(file_path)
    extension = os.path.splitext(file_name)[1].lower()
    destination_folder = file_types.get(
        extension,
        default_folder
    )
    destination_path = os.path.join(
        source,
        destination_folder,
        file_name
    )
    shutil.move(file_path, destination_path)
    print(f"Moved: {file_name} to {destination_folder}")
```

```python
#organize files
def organize_files():
    files = os.listdir(source)
    folders = set(file_types.values())
    folders.add(default_folder)
    for folder in folders:
        create_folders(os.path.join(source, folder))
    for file in files:
        file_path = os.path.join(source, file)
        if os.path.isdir(file_path):
            continue
        extensions = os.path.splitext(file_path)[1].lower()
        destination_folder = file_types.get(extensions, default_folder)
        move_file(file_path)
        with open(os.path.join(source, "file_organization_log.txt"), "a") as log_file:
            log_file.write(f"Moved: {file_path} to {destination_folder}\n")
```

```python
class FileHandler(FileSystemEventHandler):
    def on_created(self, event):
        time.sleep(1)
        move_file(event.src_path)
        
observer = Observer()
event_handler = FileHandler()
observer.schedule(
    event_handler,
    path=source,
    recursive=False
)
observer.start()
print("Watching folder...")
try:
    while True:
        time.sleep(5)
except KeyboardInterrupt:
    observer.stop()
observer.join()
```
